# List Insert

```python
#syntax: lst.insert(x, y)

lst = ['one', 'two', 'four']

lst.insert(4, "three") # will add element y at location x

print(lst)
```

# List Remove

```python
#syntax: lst.remove(x)

lst = ['one', 'two', 'three', 'four', 'two']

lst.remove('two') #it will remove first occurence of 'two' in a given list

print(lst)
```

# List Indexing

Each item in the list has an assigned index value starting from 0.

Accessing elements in a list is called indexing.

```python
lst = [1, 2, 3, 4]
print(lst[1]) #print second element

#print last element using negative index
print(lst[-2])
```

# List Slicing

Accessing parts of segments is called slicing.

The key point to remember is that the :end value represents the first value that
is not in the selected slice.

```python
numbers = [10, 20, 30, 40, 50,60,70,80]

#print all numbers
print(numbers[:])

#print from index 0 to index 3
print(numbers[0:3])
```

```python
print (numbers)
#print alternate elements in a list
print(numbers[::2])


#print elemnts start from 0 through rest of the list
print(numbers[2::2])
```

# List Comprehensions

```python
# without list comprehension
squares = []
for i in range(10):
    squares.append(i**2)   #list append
print(squares)
```

```python
#using list comprehension
squares = [i**2 for i in range(10)]
print(squares)
```

```python
#example

lst = [-10, -20, 10, 20, 50]

#create a new list with values doubled
new_lst = [i*2 for i in lst]
print(new_lst)

#filter the list to exclude negative numbers
new_lst = [i for i in lst if i >= 0]
print(new_lst)


#create a list of tuples like (number, square_of_number)
new_lst = [(i, i**2) for i in range(10)]
print(new_lst)
```

# Map and Filter

```python
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

```python
squared = []
for i in numbers:
    squared.append(i*i)
squared
```

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4, 6]
```

```python
evens = []
for i in numbers:
    if i % 2 == 0:
        evens.append(i)
evens
```

# Zip

```python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 95]

# Combine names and scores
zipped = zip(names, scores)
print(list(zipped))  # Output: [('Alice', 85), ('Bob', 90), ('Charlie', 95)]
```

```python
for name,score in zip(names,scores):
    print(f"Name: {name}, Score: {score}")
```

```python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 95, 100]

for name,score in zip(names,scores):
    print(f"Name: {name}, Score: {score}")
```
